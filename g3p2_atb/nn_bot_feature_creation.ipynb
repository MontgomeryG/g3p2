```python
# Import the required libraries
import numpy as np
import pandas as pd
import hvplot.pandas
from pathlib import Path
from pandas.tseries.offsets import DateOffset
from finta import TA
```

```python
# Read in our cleaned data
df = pd.read_csv("../resources/cleaned_data.csv")
df.head()
```

```python
df.rename(columns={"AAPL Last": "close", "AAPL Open": "open", "AAPL High": "high", "AAPL Low": "low"}, inplace=True)
```

```python
only_aapl_df = df['close', 'open', 'low', 'high']
```

```python
# Create indicators
df[("AAPL pct change")] = df[("AAPL Last")].pct_change()

# Simple Moving Averages
df["AAPL_SMA5"] = TA.SMA(df["AAPL Last"],5)
df["AAPL_SMA10"] = TA.SMA(df["AAPL Last"],10)
df["AAPL_SMA20"] = TA.SMA(df["AAPL Last"],20)
df["AAPL_SMA50"] = TA.SMA(df["AAPL Last"],50)
df["AAPL_SMA200"] = TA.SMA(df["AAPL Last"],200)

# Exponential Moving Averages
df["AAPL_EMA3"] = TA.EMA(df["AAPL Last"],3)
df["AAPL_EMA9"] = TA.EMA(df["AAPL Last"],9)
df["AAPL_EMA20"] = TA.EMA(df["AAPL Last"],20)
df["AAPL_EMA50"] = TA.EMA(df["AAPL Last"],50)
df["AAPL_EMA100"] = TA.EMA(df["AAPL Last"],100)
```
